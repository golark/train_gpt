```python
from train_gpt.model_rope import GPT, GPTConfig
import torch
model = GPT(GPTConfig(vocab_size=50304))

# load checkpoint with device-aware mapping to avoid CUDA-only deserialization errors
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
weights = torch.load('../models/rope.pt', weights_only=False, map_location=device)
w = {}
weights['model'].keys()
for k,v in weights['model'].items():
    k = k.replace('_orig_mod.', '')
    w[k] = v

model.load_state_dict(w)
```

```python
model.to_hf('./rope-hf')
```
